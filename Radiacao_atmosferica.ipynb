### Calor e equilíbrio térmico

O Calor é a transferência de energia térmica que flui de um corpo (com
maior temperatura) à outro (de menor temperatura) quando há
diferença de temperatura entre ambos. Dessa forma, o equilíbrio térmico ocorre quando os dois corpos, por meio da transferência de calor, atingem a mesma temperatura. Essa transferência pode ocorrer por condução, quando o calor se propaga através de um condutor sólido aquecido, geralmente metais, por convecção, quando substâncias ao serem aquecidas mudam de densidade criando correntes circulares, ou através da radiação emitida por todos os corpos através de ondas eletromagnéticas proporcionais a temperatura.  

### Radiação

Todos os objetos irradiam energia através de ondas eletromagnéticas chamada de Radiação Eletromagnética, essa energia é gerada pela oscilação dos elétrons do objeto e por isso quando aquecidos irradiam energia com mais intensidade do que quando frios e por isso essa mesma energia também é chamada de Radiação Térmica.

A radiação eletromagnética não precisa de um meio físico para viajar podendo ser transmitida no vácuo. Tais características fazem do sol a principal fonte de energia para a Terra e os seres vivos, além de um fator determinante para os comportamentos do tempo e do clima.

Para manter o equilíbrio térmico com a energia eletromagnética recebida do Sol, também chamada de Radiação Solar, a Terra também emite radiação chamada de Radiação Terrestre, em proporção, o que também influência nas variações das taxas de aquecimento e resfriamento do planeta. Essa radiação emitida pela Terra é igual a radiação solar absorvida ou a radiação solar incidida menos a refletida pela atmosfera.

Devido a diferença de temperatura dos corpos a radiação solar é emitida em ondas curtas até ~4 μm, já a radiação terrestre é emitida em ondas longas entre ~4 μm e ~50 μm. As ondas curtas e longas compõem o espectro eletromagnético e as radiações solar e terrestre compõe a radiação atmosférica, chamada muitas vezes apenas de radiação, que são avaliadas de forma integrada pois ambas são determinantes nas condições meteorológicas fundamentais.


### Interação da Radiação com a atmosfera

Alguns gases e elementos presentes na atmosfera são capazes de absorver a radiação solar e terrestre e como consequência do aumento e dimunição desses a distribuição da radiação na atmosfera também é afetada, sendo este um dos fatores do potencial de mudanças climáticas. Essas alterações são aferidas por um índice chamado forçante radiativa (FR) que tem sido estudada ao longo dos anos.

A Radiação também influência na circulação de elementos meteorológicos como células de conveção e temperatura, além de outros processos físicos como evaporação, biofísicos como transpiração e biológicos como fotossíntese. Ela pode ser estimada utilizando equipamentos como satélites e utilizadas para estimar fenômenos ambientais como índice de vegetação, precipitação, queimadas, entre outros, além de poderem ser assimiladas por modelos meteorológicos para prever condições dos fenômenos no tempo.

####Calculando a constante solar média (Irradiância solar média)

O Sol tem uma massa $300.000$ vezes superior a da Terra com aproximadamente $2e30$ Kg e temperatura aproximada de $6000$K em sua fotosfera. Tais parâmetros são responsáveis por uma emissão pontual de radiação solar de $6.33e7$ W/m² que devido a distância em relação atmosfera Terrestre de aproximadamente $149.6e6$ Km ou $1$ Unidade Astronômica (UA), e a lei do inverso do quadrado da distância que diz que a energia diminui sua intensidade na proporção inversa do quadrado da distância percorrida, chegam em média apenas $1366$ W/m², que é o valor da chamada constante solar ou irradiância solar total.  


---


*Lei do inverso do quadrado da distância:*
\begin{equation}
{S_0} = (\frac{E}{D^2})
\end{equation}

Onde:
* $S_0$ = Constante Solar
* $E$ = Energia do Sol
* $D^2$ = Quadrado da distância Terra-Sol


```python
# Cálcula a Constante Solar em uma determinada distância
def const_solar(distTS):
  areaSol = 4*np.pi*raioSol**2                     # Área da esfera
  energiaIni = energiaSol*areaSol                  # Energia Inicial
  energiaFim = energiaIni/(4*np.pi*distTS**2)      # Energia Final
  return energiaFim

print("Constante Solar média: %4.2f W/m²" % (const_solar(distMedTS)))
```

#### Constante solar para uma determinada distância (Irradiância solar total)

É possível calcular a constante solar para uma determinada distância utilizando uma regra de 3 simples considerando a Constante Solar Média (1366 W.m^-2) e a distância média
Terra e Sol de (149.6.10^6 Km) como referência:

\begin{equation}
{S_o} = (\frac{1}{D^2})
\end{equation}

\begin{equation}
{S_n} = (\frac{1}{D_n^2})
\end{equation}

Aplicando a regra de 3:

\begin{equation}
{S_n} = {S_o}(\frac{\overline{D}}{D_n})^2
\end{equation}



```python
def irradiancia_total(dist):
  return const_solar(distMedTS)*(distMedTS/dist)**2

print("Irrâdiancia total em Mercúrio = "+str(irradiancia_total(conv_metros(57.91e6,"km","m")))+" W/m²")
```

#### Ajuste constante solar média em função da distância Terra-Sol

Os movimentos da Terra também interferem na incidência da radiação solar. A radiação associada ao movimento de translação determina as estações do ano em cada região, já quando associada ao movimento de rotação determina o dia e a noite e o ciclo diurno da temperatura além de influenciar em processos químicos na atmosfera.

Além dos movimentos de rotação e translação, os planetas possuem particularidades como o ângulo de inclinação, excentricidade, entre outros. Uma dessas particularidades é que as órbitas dos planetas em torno do Sol são elípticas e não circulares e, portanto, a distância Terra-Sol é variável, sendo chamado de periélio o ponto em que a Terra está mais próxima do Sol e afélio o ponto mais distante. Essa diferença também influencia na quantidade de radiação incidida sobre a Terra.

A variação entre uma órbita elíptica e circular é chamada de excentricidade, seu valor varia de $0$ para uma órbita circular e aumenta à medida que a órbita se torna mais elíptica. Dos planetas do sistema solar Vênus tem a menor excentricidade com $0.0068$ e Mercúrio a maior com $0.2056$, a Terra tem excentricidade de $0.0167$.

Considerando a distância média Terra-Sol é possível ajustá-la para melhorar sua precisão. Uma das equações para este fim é a equação de Duffie e Beckman que utiliza o dia do ano, contudo dependendo da acurácia necessária existem ainda outras equações mais precisas na literatura.



---

*Equação de Duffie e Beckman:*
\begin{equation}
{D_n} = (\frac{\overline{D}}{\sqrt{1+0.033\cos(2\pi n/365)}})
\end{equation}

Onde:
* $n$ = Dia do ano
* $D_n$ = Distância Terra-Sol para o dia n (Km)
* $\overline{D}$ = Distância média Terra-Sol (Km)


```python
#Ajusta da distância Terra-Sol para um determinado dia do ano
def ajuste_dist_TS(diaDoAno):
  distTScorr = distMedTS/(np.sqrt(1+0.033*np.cos(2*np.pi*diaDoAno/365))) # Distância Terra e Sol corrigida para n em Km
  return distTScorr

#Calcula as distâncias TS para cada dia do ano
distTScorrAno = ajuste_dist_TS(diasDoAno)

## Calculo do Afério e Periélioe
distAfelio = np.max(distTScorrAno)   ## Maior distância
distPerielio = np.min(distTScorrAno) ## Menor distância

diaAfelio = int(np.where(distTScorrAno == distAfelio)[0][0])     ## Índice da maior distância TS equivalente ao dia do afélio
diaPerielio = int(np.where(distTScorrAno == distPerielio)[0][0]) ## Índice da menor distância TS equivalente ao dia do afélio

print("Afélio dia: %d, Distância Terra-Sol: %1.2e Km, Contanste solar: %4.2f W/m^2"% (diaAfelio, distAfelio, const_solar(distAfelio)))
print("Periélio dia: %d, Distância Terra-Sol: %1.2e Km, Contanste solar: %4.2f W/m^2"% (diaPerielio, distPerielio, const_solar(distPerielio)))

# Plot da curva
fig, ax = plt.subplots()
ax.plot(diasDoAno, distTScorrAno)

ax.set(xlabel='Dia do ano', ylabel='Distância em kilometros',
       title='Distância Terra-Sol')
ax.grid()

texto = ['Afélio', 'Periélio']

x_coords = [diaAfelio ,
            diaPerielio]

y_coords = [distAfelio,
            distPerielio]

for i,type in enumerate(texto):
    x = x_coords[i]
    y = y_coords[i]
    plt.scatter(x, y, color='red')
    plt.text(x+5, y+5, type, fontsize=9)

plt.show()
```

#### Declinação Solar

Devido a forma esférica dos planetas algumas áreas recebem mais radiação que outras, sendo que, a área mais próxima do Sol recebe a energia diretamente enquanto as regiões mais afastadas recebem a energia de forma obliqua distribuindo-a em uma área maior. Essa caracteristica associada ao ângulo de inclinação da Terra, próximo de $23.45$°, e ao movimento de translação são responsáveis por caracterizar as diferentes estações do ano, uma vez que ao movimentar-se um dos hemisférios se aproxima do Sol, enquanto o outro se distância. Essa angulação formada entre o plano de órbita (linha imaginária que liga o centro da Terra ao centro do Sol) e a linha do equador, é proporcional ao ângulo de inclinação da Terra variando entre $-23.45$° a $23.45$°, sendo chamado de equinócio quando o ângulo é igual a $0$ marcando o ínicio do outono e/ou primavera, dependendo do hemisfério, e solsticio quando o ângulo é igual a $±23.45$° marcando o início do inverno ou verão também dependendo do hemisfério. Essa angulação é denominada declinação solar.

---

*Declinação Solar:*
\begin{equation}
δ = 23,45°\sin[\frac{(284+n)(360/365)\pi}{180}]
\end{equation}

Onde:
* $δ$ = Declinação Solar (graus)
* $n$ = Dia do ano

```python
def declinacao_solar(diaDoAno):
  return 23.45*np.sin(np.radians((284+diaDoAno)*(360/365)))

# Calcula a declinação solar para cada dia do ano
declinacaoSolarAno = declinacao_solar(diasDoAno)

## Calcula os dias de Solstício e Equinócio

diaSolsticioMin = np.where(declinacaoSolarAno == np.min(declinacaoSolarAno))[0]
diaSolsticioMax = np.where(declinacaoSolarAno == np.max(declinacaoSolarAno))[0]

indicesNegativos = np.where(declinacaoSolarAno <= 0)[0]

diaEquinocioAntesMax = np.max(indicesNegativos[np.where(indicesNegativos <= diaSolsticioMax)])
diaEquinocioAntesMin = np.min(indicesNegativos[np.where(indicesNegativos >= diaSolsticioMax)])

print("Equinócio dias: %d e %d" % (diaEquinocioAntesMax,diaEquinocioAntesMin))
print("Solstício dias: %d e %d" % (diaSolsticioMax,diaSolsticioMin))

# Plot da curva
fig, ax = plt.subplots()
ax.plot(diasDoAno, declinacaoSolarAno)

ax.set(xlabel='Dia do ano', ylabel='Declinação solar em graus',
       title='Declinação solar ao longo do ano')
ax.grid()

texto = ['Equinócio', 'Solstício', 'Equinócio', 'Solstício']

x_coords = [diaEquinocioAntesMax ,
            diaSolsticioMax,
            diaEquinocioAntesMin,
            diaSolsticioMin]

y_coords = [declinacaoSolarAno[diaEquinocioAntesMax ],
            declinacaoSolarAno[diaSolsticioMax],
            declinacaoSolarAno[diaEquinocioAntesMin],
            declinacaoSolarAno[diaSolsticioMin]]

for i,type in enumerate(texto):
    x = x_coords[i]
    y = y_coords[i]
    plt.scatter(x, y, color='red')
    plt.text(x+0.5, y+0.5, type, fontsize=9)

plt.axhline(y=0, color='red', linestyle='--', linewidth=0.8)

plt.show()
```

#### Ângulo Horário Solar

A rotação da Terra dura em torno de $23$ horas e $56$ minutos e a translação em torno de $365$ dias e $6$ horas. Além dos movimentos de rotação e translação a Terra possui outros $12$ movimentos que também podem interferir na transferência radiativa, porém em uma escala maior.
Uma rotação completa da Terra de $360$° ocorre em $23$ horas e $56$ minutos, chamado de dia sideral, porém como a Terra também se deslocou com o movimento de translação a face que antes apontava para o sol precisa rotacionar em média mais $0.98$° ou $4$ minutos até apontar para o mesmo ponto, totalizando aproximadamente $24$ horas, o que é chamado de dia solar médio.

Na superfície terrestre um observador durante o dia tem a impressão que quem se move em torno da Terra é o Sol, nascendo no Leste e se pondo no Oeste, esse movimento é chamado de trajetória aparente do Sol, além disso durante o ano, ainda na superfície é possível observar o deslocamento dessa trajetória entre Norte e Sul, sendo que em latitudes muito altas, acima de $23.45$° ou muito baixas, abaixo de $-23.45$°, o Sol não passará exatamente acima do observador.

Considerando que a Terra demore em média $24$ horas para girar $360$° é possível concluir que ela gira em média $15$°/h, levando em conta que o Sol ao meio dia forma um ângulo de $0$ graus em relação a um observador em solo, é possível calcular ângulos solares acrescendo $15$° a cada hora solar após o meio dia e subtraindo $15$° a cada hora solar antes do meio dia, sendo assim, os ângulos solares variam de $-90$° quando o sol nasce no Leste, $0$° ao meio dia e $90$° ao se pôr no oeste.

---


*Ângulo Horário Solar:*
\begin{equation}
h = ({T_L}-12)15
\end{equation}

Onde:
* $h$ = Ângulo solar (graus)
* ${T_L}$ = Hora na longitude 0° (GMT)



```python
def angulo_horario_solar(horaGMT):
  return (horaGMT-12)*15

for horaGMT in range(6,19,1):
    print("O ângulo Horário Solar para %d horas é: %d graus" % (horaGMT, angulo_horario_solar(horaGMT)))
```

#### Ângulo Zenital


O ângulo horário solar é considerado uma aproximação pois considera a trajetória aparente do Sol perpendicular ao plano horizontal todos os dias do ano e considera também que o dia e a noite têm uma duração fixa de $12$ horas, portanto é possível deixa-la mais precisa corrigindo-a com a latitude e a declinação no ponto. Esse ângulo corrigido chama-se ângulo zenital.

---


*Ângulo Zenital:*
\begin{equation}
\cos(Z) = \cos(δ)\cos(ϕ)\cos(h)+\sin(δ)\sin(ϕ)
\end{equation}

Onde:
* $\cos(Z)$ = Cosseno do ângulo zenital (radianos)
* δ = Declinação solar (radianos)
* ϕ = Latitude (radianos)
* $h$ = Ângulo horário solar (radianos)



```python
def angulo_zenital(latitude, diaDoAno, horaGMT):
  anguloHorarioSolar = np.radians(angulo_horario_solar(horaGMT))
  declinacaoSolar = np.radians(declinacao_solar(diaDoAno))
  latitude = np.radians(latitude)
  cosZ = np.cos(declinacaoSolar)*np.cos(latitude)*np.cos(anguloHorarioSolar)+np.sin(declinacaoSolar)*np.sin(latitude)
  return np.degrees(np.arccos(cosZ))

diaDoAno = 80 #Equinócio
latitude = -50
for horaGMT in range(6,19,1):
  print("O ângulo Zenital para a posição %.2f° no dia %d às %d horas é: %.2f graus" % (latitude, diaDoAno, horaGMT, angulo_zenital(latitude, diaDoAno, horaGMT)))
```

#### Incidência da Radiação Solar na Superfície

Considerando um ponto na superfície, e negligenciando a interação da radiação solar com a atmosfera é possível calcular a quantidade de energia solar que incide neste ponto.

---


*Incidência de radiação na superfície (potencial):*
\begin{equation}
S = {S_o}\cos(Z)
\end{equation}

Onde:
* S = Radiação incidente na superfície ignorando a interação com a atmosféra
* ${S_o}$ = Radiação incidente no topo da atmosféra
* $\cos(Z)$ = Cosseno do ângulo zenital (radianos)


```python
def radiacao_sfc_sem_atmosfera(constSolar, Z):
  return constSolar*np.cos(np.radians(Z))

diaDoAno = 80 #Equinócio
latitude = 0
for horaGMT in range(6,19,1):
  constSolar = const_solar(distMedTS)
  Z = angulo_zenital(latitude, diaDoAno, horaGMT)
  print("A Radiação incidente na sfc para a posição %.2f° no dia %d às %d horas é: %.2f W/m^2" % (latitude, diaDoAno, horaGMT, radiacao_sfc_sem_atmosfera(constSolar, Z)))
```

#### Ângulo Horário Máximo

Com o ângulo Zenital é possível saber o potencial energético de um ponto em determinado momento do ano e do dia, isso é importante para questões de energia renováveis, precaver de insolação, entre outros.

Trocando alguns parâmetros da equação do ângulo zenital é possível obter informações por exemplo do ângulo horário ao amanhecer e entardecer em uma determinada latitude:

---


*Ângulo horário Máximo:*
\begin{equation}
H = |\cos^{-1}(\tan(δ)\tan(ϕ)|
\end{equation}

Onde:
* H = Ângulo horário máximo (radianos)
* δ = Declinação solar (radianos)
* ϕ = Latitude (radianos)

```python
def angulo_horario_max(latitude, diaDoAno):
  declinacaoSolar = np.radians(declinacao_solar(diaDoAno))
  latitude = np.radians(latitude)
  H = np.tan(declinacaoSolar)*-1*np.tan(latitude)
  return np.degrees(np.arccos(H))

# Rio de Janeiro
latitude = 22.9068

#Equinócio dias: 80 e 263
#Solstício dias: 171 e 354
print("Ângulo Solar máximo no Rio de Janeiro no Sólsticio de Verão: %.2f°" % angulo_horario_max(latitude, 171))
print("Ângulo Solar máximo no Rio de Janeiro no Equinócio: %.2f°" % angulo_horario_max(latitude, 80))
print("Ângulo Solar máximo no Rio de Janeiro no Sólsticio de Inverno: %.2f°" % angulo_horario_max(latitude, 354))
```

#### Fotoperíodo

Com o ângulo correto de nascer e por do Sol e sabendo que a Terra gira $15$°/h é possível calcular a duração em horas solares do dia chamado de dia astronômico ou fotoperiodo(T):

---


*Fotoperíodo:*
\begin{equation}
T = \frac{2H}{15}
\end{equation}

Onde:
* T = Fotoperíodo (horas)
* H = Ângulo horário máximo (radianos)


```python
def fotoperiodo(latitude, diaDoAno):
  return 2*angulo_horario_max(latitude, diaDoAno)/15

latitude=51.5 #Londres

print("Duração do dia em Londres no Sólsticio de Inverno: %.2f horas" % fotoperiodo(latitude, 171))
print("Duração do dia em Londres no Equinócio: %.2f horas" % fotoperiodo(latitude, 80))
print("Duração do dia em Londres no Sólsticio de Verão: %.2f horas" % fotoperiodo(latitude, 354))
```

#### Nascer e pôr do Sol

Definido a duração do dia astronômico é possível calcular também o horário correto do nascer e pôr do Sol, sabendo que 12 horas é meio dia basta subtrair metade do fotoperíodo para decobrir o horário do nascer do Sol e o oposto para conhecer o horário do pôr do Sol:

---

*Nascer do Sol:*
\begin{equation}
{N_s} = 12-\frac{T}{2}
\end{equation}

*Pôr do Sol:*
\begin{equation}
{P_s} = 12+\frac{T}{2}
\end{equation}

Onde:
* T = Fotoperíodo (horas)
* ${N_s}$ = Nascer do Sol (horas)
* ${P_s}$ = Pôr do Sol (horas)

```python
def hora_nascer_sol(latitude, diaDoAno):
  return 12-fotoperiodo(latitude, diaDoAno)/2

def hora_por_sol(latitude, diaDoAno):
  return 12+fotoperiodo(latitude, diaDoAno)/2

print("Nascer do Sol em Londres no Sólsticio de Verão: %.2f horas" % hora_nascer_sol(latitude, 354))
print("Pôr do Sol em Londres no Sólsticio de Verão: %.2f horas" % hora_por_sol(latitude, 354))
```

#### Equação do Tempo

Devido a forma eliptica da órbita terrestre a velocidade de translação não é constante e dependendo da época do ano a velocidade pode ser maior ou menor, com a equação do tempo é possível calcular a diferença de tempo em minutos no momento em que o sol passa pelo meridiano em determinada época e utiliza-la para corrigir outras medidas.

---

*Equação do Tempo:*
\begin{equation}
{E_t} = 229.1[0.000075+0.001868\cos(Θ)-0.032077\sin(Θ)-0.014615\cos(2Θ)-0.040849\sin(2Θ)]
\end{equation}

Onde:
* ${E_t}$ = Correção do tempo em que o Sol cruza o meridiano (minutos)
* Θ = $\frac{2\pi(n-1)}{365}$ (radianos)
* n = dia do ano (1 a 365/366)


```python
def equacao_tempo(diaDoAno):
  teta = (2*np.pi*(diaDoAno-1))/365
  ET = 229.1*(0.000075+0.001868*np.cos(teta)-0.032077*np.sin(teta)-0.014615*np.cos(2.*teta)-.040849*np.sin(2.*teta))
  return ET

# Horário em que o Sol cruza o meridiano:
meridiano = equacao_tempo(diasDoAno)

# Plot da curva
fig, ax = plt.subplots()
ax.plot(diasDoAno, meridiano)

ax.set(xlabel='Dia do ano', ylabel='Cruza o meridiano',
       title='Equação do Tempo')
ax.grid()

plt.axhline(y=0, color='red', linestyle='--', linewidth=0.8)

plt.show()
```

#### Hora Solar

Por fim, com base no movimento aparente Leste-Oeste, dependendo da posição longitudinal do observador o Sol pode estar mais próximo ou mais afastado, modificando o ângulo zenital, sendo assim como o movimento completo de translação dura em média 24 horas, são 15° a cada hora ou 1° a cada 4 minutos. Com isso e com a equação do tempo é possível calcular a hora solar de uma determinada coordenada geográfica.  

---

*Hora solar:*
\begin{equation}
{H_S}={T_L}+4δL+{E_t}
\end{equation}

Onde:
* ${H_S}$ = Hora Solar (minutos)
* ${T_L}$ = Hora na longitude 0° (minutos)
* δL = Longitude (graus)
* ${E_t}$ = Equação do Tempo (minutos)

```python
def hora_solar(horaGMT, longitude, diaDoAno):
  return ((horaGMT*60)+4*longitude+equacao_tempo(diaDoAno))/60
```

#### Incidência da Radiação Solar no topo da atmosferá para uma determinada data e localização

Radiação solar no topo para uma data, horário e local (Irradiância):

\begin{equation}
{S} = {S_0}(\frac{\overline{D}}{D_n})^2{cosZ}
\end{equation}

Radiação no topo da atmosfera recebida ao longo de um dia (Irradiação):

\begin{equation}
{Q_o} = \int_{tn}^{tp} {S_0}(\frac{\overline{D}}{D_n})^2{cosZ}\;dt
\end{equation}

Decompondo a função ela resulta na função para locais com nascer e por do Sol:

\begin{equation}
{Q_o} = 37,6.10^6(\frac{\overline{D}}{D_n})^2[\cos(\delta)\cos(\varphi)\cos(h)+\sin(\delta)\sin(\varphi)h]
\end{equation}

E na função para os locais sem nascer e por do Sol:

\begin{equation}
{Q_o} = \frac{2\pi{S_0}}{\omega}(\frac{\overline{D}}{D_n})^2[\sin(\delta)\sin(\varphi)]
\end{equation}

Onde:
* ${S_0}$ = Constante solar (W/m²)
* $\overline{D}$ = Distância média Terra-Sol (metros)
* ${D_n}$ = Distância Terra-Sol no dia n (metros)
* $\cos(Z)$ = Cosseno do ângulo zenital (radianos)
* δ = Declinação solar (radianos)
* ϕ = Latitude (radianos)
* $h$ = Ângulo horário solar (radianos)
* $\omega$ = $\frac{2.\pi}{24*60*60}$ = 7.27199e-05 (radianos/segundos)


```python
def irrad_diaria_topo_com_por_sol(latitude, dia_do_ano):
  H = angulo_horario_max(latitude, dia_do_ano)
  distTS = ajuste_dist_TS(dia_do_ano)
  declinacaoSolar = declinacao_solar(dia_do_ano)
  irrd = (37.6*10**6)* \
          ((distMedTS/distTS)**2)*\
          (np.cos(np.radians(declinacaoSolar))*np.cos(np.radians(latitude))*np.sin(np.radians(H))+\
           np.sin(np.radians(declinacaoSolar))*np.sin(np.radians(latitude))*np.radians(H))
  return irrd/1000000

def irrad_diaria_topo_sem_por_sol(latitude, dia_do_ano):
  distTS = ajuste_dist_TS(dia_do_ano)
  constSolar = const_solar(distTS)
  declinacaoSolar = declinacao_solar(dia_do_ano)
  omega = 7.27199e-05
  irrd = ((2*np.pi*constSolar)/omega)* \
          ((distMedTS/distTS)**2)*\
          (np.sin(np.radians(declinacaoSolar))*np.sin(np.radians(latitude)))
  #irrd = np.sqrt(irrd**2)
  return irrd/1000000
```

#### Dados observacionais de radiação solar

Os diversos instrumentos utilizados para medir radiação em superficie são chamados genericamente de Radiômetros, eles tem diversas aplicações como estudar os constituíntes atmosféricos, as propriedades da superfície e da atmosfera, o balanço de energia, mudanças climáticas, validar modelos, entre outros. Também podem ser configurados para medir em uma região espectral específica dependendo da aplicabilidade, o ultravioleta por exemplo pode ser utilizado para análise e monitoramento do ozônio e outros tipos de poluição enquanto o infravermelho para mudanças climáticas.
Os radiometros são instrumentos capazes de medir um ou vários tipos tipos de radiação entre elas:
* Radiação direta (Rd): proveniente do topo da atmosfera diretamente sobre o sensor ( sofre redução conforme o ângulo zenital, concentração de gases, partículas de aerossóis, nuvens, etc
* Radiação difusa (Rdif): radiação proveniente do espalhamento por gases, partículas de aerossóis, nuvens, etc, que acabam incidindo sobre o sensor.
* Radiação solar Global (Rglo): Radiação solar
direta + difusa (Rglo= Rd + RDif)
* Radiação solar refletida pela sfc (Rr): Proveniente da Rglo incidida na superficie e que não foi absorvida, para medida dessa radiação os sensores são voltados para baixo.

Dentre os principais tipos de radiômetros estão o Piranômetro, Pereliômetro, Espectroradiômetro/Espectrofotômetro, Fotômetro solar, Heliográfo, Ceilômetro, entre outros.
Os sensores devem ser instalados e calibrados conforme recomendações da Organização Mundial Meteorológica(OMM), na qual especificam as melhores condições para o funcionamento adequado dos sensores.

Alguns desses sensores medem a insolação que é o período do dia em que houve incidencia de radiação sem obstrução, por nuvens ou outros componentes.

As redes desses equipamentos geralmente são mantidas por organizações governamentais e distribuidas publicamente, contudo é preciso atentar-se para a qualidade dos dados, observando a precisão, exatidão e suas incertezas.

No Brasil instituições como o INMET e o SONDA fornecem dados de radiometros.

### Espectro eletromagnético

Durante muito tempo, a luz era a única parte conhecida do espectro eletromagnético. Os gregos antigos tinham a noção de que a luz viajava a forma de linhas retas, chegando a estudar algumas de suas propriedades, que fazem parte do que atualmente denominamos óptica geométrica. Foi somente nos séculos XVI e XVII que o estudo da luz passou a gerar teorias conflitantes quanto a sua natureza.

A primeira descoberta de ondas eletromagnéticas além da luz ocorreu em 1800, quando William Herschel descobriu a radiação infravermelha. Em seu experimento, Herschel direcionou a luz solar através de um prisma, decompondo-a, e então mediu a temperatura de cada cor. Ele descobriu que a temperatura aumentava do violeta para o vermelho, e que a temperatura mais alta se encontrava logo após o vermelho, numa região em que nenhuma luz solar era visível.

No ano seguinte, Johann Wilhelm Ritter realizou estudos na outra ponta do espectro visível e percebeu a existência do que ele chamou de "raios químicos" (raios de luz invisíveis que provocavam reações químicas), que se comportavam de forma semelhante aos raios de luz violeta visíveis, mas que estavam além deles no espectro. O termo "raios químicos" foi posteriormente renomeado radiação ultravioleta.

A radiação eletromagnética foi pela primeira vez relacionada com o eletromagnetismo em 1845, quando Michael Faraday percebeu que a direção de polarização da luz que passava por um material transparente respondia a um campo magnético. Esse fenômeno foi mais tarde denominado Efeito Faraday. Durante a década de 1860, James Maxwell mostrou que, a partir das equações de Maxwell, era possível encontrar uma equação de onda para descrever a propagação do campo elétrico e outra para o campo magnético. Analisando a velocidade dessas ondas do ponto de vista teórico, Maxwell descobriu que elas deviam viajar à velocidade da luz, o que o levou a inferir que a própria luz deveria ser uma onda eletromagnética.

As equações também previam um número infinito de frequências para as ondas eletromagnéticas, todas elas viajando à velocidade da luz. Esse foi o primeiro indício da existência que um espectro eletromagnético completo.

Em 1895 Wilhelm Röntgen percebeu um novo tipo de radiação emitida durante um experimento com um tubo com vácuo sujeito à alta voltagem. Ele chamou essa radiação de raios-x e descobriu que eles eram capazes de atravessar partes do corpo humano mas eram refletidos ou parados por materiais densos, como os ossos, e passaram a ser amplamente usados na medicina.

A última porção do espectro eletromagnético foi completado com a descoberta dos raios gama. Em 1900 Paul Villard estava estudando as emissões radiativas do radium quando ele identificou um novo tipo de radiação que ele primeiramente pensou se tratar de partículas semelhantes às conhecidas partículas alfa e beta, mas com a propriedade de serem bem mais penetrantes que ambas.

Entretanto, em 1910 o físico William Henry Bragg demonstrou que os raios gama eram uma radiação eletromagnética, e não partícula, e em 1914, Ernest Rutherford (que havia nomeado a radiação de raios gamas em 1903 quando percebeu que eles eram fundamentalmente diferentes de partículas alfa e beta) e Edward Andrade mediram seus comprimentos de onda e descobriram que os raios gama eram semelhantes ao raio-x, porém com comprimentos menor e maior frequência.

#### Ondas Eletromagnéticas

* Se propagam afastando-se da fonte
* Idenpende de meio para se propagar
* Muda de velocidade de acordo com o meio de propagação, fazendo-a mudar de direção
* No vácuo viaja na velocidade da luz $3e8$ m/s (c)
* Carrega energia
* Pode ser expressa em comprimento de onda (λ), equivalente a distância em metros, entre 2 picos ou 2 vales de uma onda
* Pode ser expressa em frequência (Hz/s), equivalente ao número de λ que passam por um ponto em um segundo
* São polarizadas
* Variação (ou não) da intensidade com a direção (isotrópica/não isotrópica)

É possível descrever as ondas em relação à porção do espectro como:

* monocromática: Um único comprimento de onda
* broadband: Um intervalo de comprimento de ondas  

É possível descrever as ondas em relação à polarização como:

* coerente: Quando produzida a partir de fontes artificiais como radares, torres de rádio e TV, forno microondas, etc. Essas podem ser de tipos diferentes como linear, circular ou elíptica
* incoerente: Produzida por fontes naturais. Na atmosfera a radiação emitida é não polarizada mas esta pode tornar-se polarizada ao interagir com superfícies ou partículas

Sabendo que todas as ondas viajam na mesma velocidade (c) é possível calcular o comprimento de onda (λ) dividindo a velocidade pela frequência (v) ou calcular a frequência dividindo a (c) por (λ)

---

*Comprimento de onda:*
\begin{equation}
λ=\frac{c}{\tilde{v}}
\end{equation}

*frequência:*
\begin{equation}
\tilde{v}=\frac{c}{λ}
\end{equation}

Onde:
* λ = Comprimento de onda (metros)
* c = velocidade da luz (m/s)
* $\tilde{v}$ = frequência (Hz/s)



```python
def comprimento_onda(frequencia):  ## Recebe em Hz e retorna em µm
  return conv_metros(velLuzVacuo/frequencia,"m","um")

def frequencia(comprimentoOnda):  ## Recebe em metros e retorna em Hz
  return velLuzVacuo/comprimentoOnda


print("{:.4e}".format(conv_hertz(frequencia(conv_metros(0.01, "um", "m")),"Hz","GHz"))+"GHz = 0.01µm")
print("{:.4e}".format(conv_hertz(frequencia(conv_metros(0.65, "um", "m")),"Hz","GHz"))+"GHz = 0.65µm")
print("{:.4e}".format(conv_hertz(frequencia(conv_metros(11.2, "um", "m")),"Hz","GHz"))+"GHz = 11.2µm")

print("183GHz = "+"{:.4e}".format(comprimento_onda(conv_hertz(183,"GHz","Hz")))+"µm")
print("89GHz = "+"{:.4e}".format(comprimento_onda(conv_hertz(89,"GHz","Hz")))+"µm")
print("19GHz = "+"{:.4e}".format(comprimento_onda(conv_hertz(19,"GHz","Hz")))+"µm")
```

A radiação eletromagnética também é expressa em função do número de onda, que é uma grandeza física inversamente proporcional ao comprimento de onda e pode ser definido como o número de comprimentos de onda por unidade de distância, podendo ser calculada em função do comprimento de onda ou da frequência.

---

*Número de onda em função do comprimento de onda:*
\begin{equation}
v=\frac{1}{λ}
\end{equation}

*Número de onda em função da frequência:*
\begin{equation}
v=\frac{\tilde{v}}{c}
\end{equation}

Onde:
* λ = Comprimento de onda (metros)
* c = velocidade da luz (m/s)

#### Energia e Fóton

As ondas eletromagnéticas são capazes de transportar energia (radiação).

A radiação não atinge uma superfície como um fluxo contínuo e suave de energia, mas sim em "grupos" de partículas discretos chamados fotóns, descritos por Planck(1900)

A constante de Planck é importante para determinar a energia de um fóton, o que é obtido mediante a seguinte equação:

---

*Quanto maior a frequência de uma radiação, maior sua energia*
\begin{equation}
E=hv
\end{equation}

Onde:
* E: energia (J)
* h: constante de Planck = 6.626e-34 Joules
* v: frequência das radiações eletromagnéticas (Hz) = c/λ (velocidade da luz/comprimenro de onda)

```python
def energia_do_foton_pelo_co(comprimentoOnda):
  return constantePlanck*velLuzVacuo*100/comprimentoOnda

def energia_do_foton_pela_freq(frequencia):
  return constantePlanck*frequencia

print("{:e}".format(energia_do_foton_pelo_co(11.2)))
print("{:e}".format(energia_do_foton_pela_freq(conv_hertz(2.6786,"GHz","Hz"))))
```

A menor energia de um fóton é $h.v$, assim, todos os valores de energia devem ser múltiplos e inteiros desse valor, o que significa dizer que não temos iaguais a números fracionados como $0,6.h.v$ mas inteiros como $2.h.v$, $3.h.v$ e assim por diante. Essa energia pode ser também representada como:

---

*Quanto maior a frequência de uma radiação, maior sua energia*
\begin{equation}
E=nhv
\end{equation}

Onde:
* E: energia (J)
* n: número inteiro
* h: constante de Planck = 6.626e-34 Joules
* v: frequência das radiações eletromagnéticas (Hz) = c/λ (velocidade da luz/comprimenro de onda)



#### Fluxo/Irradiância

Representa o fluxo de energia irradiada (W) ou (J/s) que incide sobre uma superfície, por unidade de área (m²), ou seja, integral da energia (J) vinda de todas as direções.

* Densidade de fluxo (F) é a forma de descrever o transporte de energia (W/m²) também chamada de irradiância.
* Magnitude do fluxo depende da orientação da superfície emissora/incidente (COZ)
* Intensidade do fluxo (I) depende da quantidade de energia emitida e da direção, além de fatores atmosféricos como presença de nuvens

* Fluxo deve ser uma quantitizado em uma faixa de banda já que o fluxo de um único comprimento de onda é muito baixo

O fluxo de radiação eletromagnética, ao se propagar pelo espaço, pode interagir com a superfície ou objetos, sendo por estes absorvido, refletido, ou reemitido.

---

*Fluxo monocromático ou espectral:*
\begin{equation}
{F_λ}=\lim_{Δ_λ\to0} F\frac{(λ ,λ+{Δ_λ})}{{Δ_λ}}
\end{equation}

*Fluxo de banda larga:*
\begin{equation}
F({λ_1},{λ_2})=\int_{λ_1}^{λ_2} F(λ)\; dλ
\end{equation}

Onde:
* λ = comprimenro de onda (metros)

Assim como podemos entender um ângulo plano em radiano como um setor da circunferencia de um círculo dividido pelo raio, podemos entender um ângulo sólido como um "setor" de uma esfera, definido pela razão entre o elemento de área na superfície da esfera e o seu raio ao quadrado.

Assim o maior ângulo plano é aquele que subtende toda a circunferência do círculo pelo raio, ou 2πr/r = 2π, o maior ângulo sólido subtende toda a área superficial da esfera, e vale 4πr²/r² = 4π esferorradianos/sterradian (sr).

* Uma esfera tem ângulo sólido Ω = 4π sr

A irradiancia é a quantidade de energia eletromagnética (como luz) que incide em uma determinada área de superfície, medida em watts por metro quadrado (W/m²). Já a radiância é a quantidade de energia eletromagnética emitida ou refletida de uma superfície em todas as direções, medida em watts por metro quadrado por esteradiano (W/m²/sr).

#### Radiação Térmica e Corpo Negro

Todos os corpos quando aquecidos acima do zero absoluto emitem radiação eletromagnética, que por estar ligada a temperatura, também é chamada de radiação térmica. Como na natureza todos os corpos estão em uma determinada temperatura, todos eles emitem esse tipo de radiação. Essa radiação quando incidida em outros corpos com temperatura inferior é parcialmente absorvida e parcialmente refletida, aumentando a temperatura do corpo mais frio até ambos estarem em equilíbrio térmico. Devido a essas interações entre os corpos é impossível aferir com exatidão na natureza o total de radiação emitida por um corpo a uma determinada temperatura.

O corpo negro foi idealizado como sendo um corpo opaco que absorve toda radiação que incide sobre ele, ou seja, nenhuma parte da radiação é refletida ou transmitida, sendo assim em equilíbrio térmico ele emite energia na mesma proporção em que absorve tornando possível aferir a radiação emitida a uma determinada temperatura.

Apesar de ser um modelo ideal, o corpo negro não é encontrado na natureza e embora seja uma abstração teórica teve uma importância fundamental pois para algumas faixas do espectro, muitos corpos reais se comportam como se fossem um corpo negro, tornando possível a estimativa do espectro de emissão para diferentes frequências e mais tarde a formulação de teorias como a Lei de Wien e a a Lei de Rayleigh-Jeans, que apesar de não responderem idealmente ao problema contribuíram para que Planck por fim idealiza-se uma função ideal.



---

*Função de Planck*
\begin{equation}
{B_λ}(T)=\frac{2hc^2}{λ^5(e^\frac{hc}{{k_b}λT})-1}
\end{equation}

Onde:
* $B_λ(T)$ = Energia emitida por um corpo em uma determinada temperatura e comprimento de onda (W/m²/µm/sr)
* $λ$ = Comprimento de onda (μm)
* $T$ = Temperatura superficial do corpo (K)
* $c$ = Velocidade da luz no vácuo (m/s)
* $h$ = Constante de Planck = 6.626e-34 (J.s)
* ${K_b}$ = Constante de Boltzmann = 1.381e-23 (J/K)



```python
def planck(T, co):  # T em Kelvin, co em µm e output em W/m^2/μm/sr
  c1=(2*constantePlanck*velLuzVacuo**2)*(conv_metros(1,"m","um")**4)          # W/m^2/μm^4/sr
  c2=(constantePlanck*conv_metros(velLuzVacuo,"m","um"))/(constanteBoltzmann*co*T)
  return (c1/co**5.)*1/(np.exp(c2)-1)

# Humano
print("Humano a 309K no comprimento de onda 0.5µm emite:"+"{:.3e}".format(planck(309,0.5))+" W/m²/µm/sr")
print("Humano a 309K no comprimento de onda 3.9µm emite:"+"{:.3e}".format(planck(309,3.9))+" W/m²/µm/sr")
print("Humano a 309K no comprimento de onda 11.0µm emite:"+"{:.3e}".format(planck(309,11.0))+" W/m²/µm/sr")
print("")
# Queimadas
print("Queimadas a 800K no comprimento de onda 0.5µm emite:"+"{:.3e}".format(planck(800,0.5))+" W/m²/µm/sr")
print("Queimadas a 800K no comprimento de onda 3.9µm emite:"+"{:.3e}".format(planck(800,3.9))+" W/m²/µm/sr")
print("Queimadas a 800K no comprimento de onda 11.0µm emite:"+"{:.3e}".format(planck(800,11.0))+" W/m²/µm/sr")
print("")
# Sol
print("Sol a 6000K no comprimento de onda 0.5µm emite:"+"{:.3e}".format(planck(6000,0.5))+" W/m²/µm/sr")
print("Sol a 6000K no comprimento de onda 3.9µm emite:"+"{:.3e}".format(planck(6000,3.9))+" W/m²/µm/sr")
print("Sol a 6000K no comprimento de onda 11.0µm emite:"+"{:.3e}".format(planck(6000,11.0))+" W/m²/µm/sr")
```

####Função de Planck em função da frequência ($υ$)

Derivando a Função de Planck em função do comprimento de onda para função da frequência:

---

\begin{equation}
{B_λ}(T)=\frac{2hc^2}{λ^5(e^\frac{hc}{{k_b}λT})-1}
\end{equation}

---

\begin{equation}
{B_v}\frac{dv}{d_λ}(T)=\frac{2hc^2}{λ^5(e^\frac{hc}{{k_b}λT})-1}
\end{equation}

---

\begin{equation}
{B_v}\frac{c}{λ^2}(T)=\frac{2hc^2}{λ^5(e^\frac{hc}{{k_b}λT})-1}
\end{equation}

---

\begin{equation}
{B_v}(T)=\frac{2hc}{λ^3(e^\frac{hc}{{k_b}λT})-1}
\end{equation}

---

\begin{equation}
{B_v}(T)=\frac{2hc}{(\frac{c}{v})^3(e^\frac{hc}{{k_b}(\frac{c}{v})T})-1}
\end{equation}

---

\begin{equation}
{B_v}(T)=\frac{2hcv^3}{c^3(e^\frac{hcv}{{k_b}cT})-1}
\end{equation}

---

\begin{equation}
{B_v}(T)=\frac{2hv^3}{c^2(e^\frac{hv}{{k_b}T})-1}
\end{equation}

Onde:
* ${B_v(T)}$ = Energia emitida por um corpo em uma determinada temperatura e frequência (W/m²/Hz)
* $v$ = Frequência (Hz)
* $λ$ = Comprimento de onda (μm)
* $T$ = Temperatura superficial do corpo (K)
* $c$ = Velocidade da luz no vácuo (m/s)
* $h$ Constante de Planck = 6.626e-34 (J.s)
* ${K_b}$ = Constante de Boltzmann = 1.381e-23 (J/K)


```python

def planck_f(T, f):  # T em Kelvin, f em Hz e output em W/m^2/Hz
  return (2*constantePlanck*f**3/velLuzVacuo**2)*1/(np.exp(constantePlanck*f/(constanteBoltzmann*T))-1)


ondas = np.arange(100,2100,100) #nm
temps = np.arange(3500,6000,500) #K
freqs = velLuzVacuo/conv_metros(ondas, "nm", "m") #Hz

fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)

for T in temps:
  energ = planck_f(T, freqs)
  ax.plot(freqs, energ*freqs, label=T)

ax.legend(loc=1, title = "Temperatura (K)")
ax.set(xlabel='Hz', ylabel='W/m^2/µm/sr')

print(energ*freqs)
plt.xticks(rotation=90)

ax.set_xticks(freqs)

ax.grid()

plt.show()
```

#### Lei de Wien

É possível estimar a temperatura máxima de uma fonte a partir do seu comprimento de onda e vice-versa.

se comparar a energia ${E_λ}$ em duas temperaturas diferentes, não no mesmo comprimento de onda, mas em comprimentos de onda inversamente proporcionais a $T$, então os valores de ${E_λ}$ serão proporcionais a quinta potência da temperatura absoluta.

Essa relação foi proposta teoricamente por Wien antes da dedução da função de Planck, contudo é possível obter a mesma função calculando a derivada da função de Planck igual a zero: ${B_λ}(T)' = 0$.

---

*Lei de Wien*
\begin{equation}
{λ_{max}}=\frac{C}{T}
\end{equation}

Onde:
* ${λ_{max}}$: Comprimento de onda máximo para uma determinada temperatura (metros)
* C : constante de proporcionalidade = 2976 µm.K
* T = Temperatura superficial do corpo (K)


```python
def lei_de_wien(T):
  return 2976/T

print("Comprimento de onda de máxima emissão para um ser humano a 35.5°C = "+"{:2.3f}".format(lei_de_wien(35.5+273.15))+"µm")
print("Comprimento de onda de máxima emissão para o Planeta Terra a 20°C = "+"{:2.3f}".format(lei_de_wien(20+273.15))+"µm")
print("Comprimento de onda de máxima emissão para o Sol a 6000K = "+"{:2.3f}".format(lei_de_wien(6000))+"µm")
print("Comprimento de onda de máxima emissão para uma fogueira a 500°C = "+"{:2.3f}".format(lei_de_wien(500+273.15))+"µm")
```

#### Lei de Stefan-Boltzmann

Também foi proposta teoricamente antes da Função de Planck utilizando apenas os conceitos de radiação de corpo negro e termodinâmica onde diz que a quantidade máxima de energia que pode ser emitida por um corpo negro é
proporcional a quarta potência da temperatura absoluta desse corpo. Seria o equivalente a integrar a função de Planck.

---

*Lei de Stephan-Boltzmann*
\begin{equation}
{E_{max}}=δ.T^4
\end{equation}

Onde:
* ${E_{max}}$: Máximo de energia emitida por um corpo em uma determinada temperatura (W.m²)
* δ : Constante de Steffan = 5.6696e-8 W/m²/K^4
* T = Temperatura superficial do corpo (K)


```python
def lei_de_boltzmann(T):
  return (constanteStefan)*T**4

print("Radiação emitida pelo Sol considerando-o como um corpo negro a 5760 K = "+"{:.3e}".format(lei_de_boltzmann(5760))+" W.m^-2")
```

#### Emissividade Monocromática

É a razão entre o que é emitido e o que seria emitido se o corpo se  comportasse com um corpo negro ${ϵ_λ}$.

Quando ${ϵ_λ}$=1, o corpo é efetivamente um corpo negro para esse comprimento de onda.

---

*Emissividade monocromática*
\begin{equation}
{ϵ_λ}=\frac{E_λ}{{B_λ}(T)}
\end{equation}

Onde:
* ${ϵ_λ}$: Emissividade (0 <= ${ϵ_λ}$ <= 1)
* ${E_λ}$: Energia emitida por um corpo em um determinado comprimento de onda e temperatura
* ${{B_λ}(T)}$ = Energia emitida por um corpo negro em um determinado comprimento de onda e temperatura



#### Emissividade de um corpo cinza

Emissividade de Corpo Cinza: É a razão entre o fluxo de banda larga F(λ1,λ2)
observado e aquele previsto pela lei de Stefan-Boltzmann

---

*Emissividade de corpo cinza*
\begin{equation}
{ϵ}=\frac{F(λ1,λ2)}{E_{max}}
\end{equation}

Onde:
* ${ϵ}$: Emissividade (0 <= ${ϵ_λ}$ <= 1)
* $F(λ1,λ2)$: Fluxo de banda larga
* ${E_{max}}$ = Máximo de energia emitida por um corpo em uma determinada temperatura (W.m²)

#### Lei de Kirchhoff da radiação térmica

A lei de Kirchhoff iguala a emissão e a absorção em objetos aquecidos, ela é amplamente aplicada nos problemas de radiação atmosférica, porém, ela só é válida sob condição de Equilíbrio Termodinâmico Local (ETL) no qual as moléculas trocam energia através de colisões muito mais rapidamente do que por campo de radiação ou outras formas de energia

---

*Em equilíbrio térmico, a emissividade de um corpo é igual à sua absortância*
\begin{equation}
{ϵ_λ}(Θ, ϕ)={a_λ}(Θ, ϕ)
\end{equation}

Onde:
* ${ϵ_λ}(Θ, ϕ)$: Emissividade
* ${a_λ}(Θ, ϕ)$ = Absortância


#### Temperatura de Brilho

É a relação entre Intensidade da radiação medida em um certo comprimento de onda e a temperatura que emitiria essa Intensidade.

---

*Temperatura de brilho*
\begin{equation}
{T_B} = \frac{C2}{λ\ln{(1+\frac{C1}{λ^5.{B_λ}})}}
\end{equation}

Onde:
  c1=2*constantePlanck*velocidadeLuz**2
  c2=constantePlanck*velocidadeLuz/constanteBoltzmann

* C1: $2.h.c^2$ (h = constante de Planck, c = velocidade da luz)
* C2: $h.\frac{c}{δ}$ (h = constante de Planck, c = velocidade da luz, δ = constante de Boltzmann)
* ${B_λ}$: Radiação emitida
* λ: Comprimento de onda


```python
def temp_brilho_co(R, co):  # R em W/m^2/μm/sr, co em µm e output em Kelvin
  c1=(2*constantePlanck*velLuzVacuo**2)*(conv_metros(1,"m","um")**4)          # W/m^2/μm^4/sr
  c2=constantePlanck*conv_metros(velLuzVacuo,"m","um")/constanteBoltzmann
  return c2/(co*np.log(1+c1/(co**5.*R)))

# Humano
print("Humano emitindo 1.32e-31 W/m^2/µm/sr no comprimento de onda 0.5µm está a:"+"{:.3e}".format(temp_brilho_co(1.32e-31,0.5))+"K")
print("Humano emitindo 8.588e-01 W/m^2/µm/sr no comprimento de onda 3.9µm está a:"+"{:.3e}".format(temp_brilho_co(8.588e-01,3.9))+"K")
print("Humano emitindo 1.088e+01 W/m^2/µm/sr no comprimento de onda 11.0µm está a:"+"{:.3e}".format(temp_brilho_co(1.088e+01,11.0))+"K")
print("")
# Queimadas
print("Queimadas emitindo 8.988e-07 W/m^2/µm/sr no comprimento de onda 0.5µm está a:"+"{:.3e}".format(temp_brilho_co(8.988e-07,0.5))+"K")
print("Queimadas emitindo 1.324e+03 W/m^2/µm/sr no comprimento de onda 3.9µm está a:"+"{:.3e}".format(temp_brilho_co(1.324e+03,3.9))+"K")
print("Queimadas emitindo 1.792e+02 W/m^2/µm/sr no comprimento de onda 11.0µm está a:"+"{:.3e}".format(temp_brilho_co(1.792e+02,11.0))+"K")
print("")
# Sol
print("Sol emitindo 3.173e+07 W/m^2/µm/sr no comprimento de onda 0.5µm está a:"+"{:.3e}".format(temp_brilho_co(3.173e+07,0.5))+"K")
print("Sol emitindo 1.555e+05 W/m^2/µm/sr no comprimento de onda 3.9µm está a:"+"{:.3e}".format(temp_brilho_co(1.555e+05,3.9))+"K")
print("Sol emitindo 3.039e+03 W/m^2/µm/sr no comprimento de onda 11.0µm está a:"+"{:.3e}".format(temp_brilho_co(3.039e+03,11.0))+"K")
print("")
```

### Lei de Beer

Trata sobre a transferência radiativa em um determinado caminho geométrico

#### Processos de Interação radiativa

Parte da energia incidida ${E_{inc}}$ sobre um corpo "não opaco" em um determinado comprimento de onda é refletida ${E_r}$, parte é absorvida ${E_a}$ e parte é transmitida ${E_t}$, consequentemente a soma dessas partes deve ser igual a energia incidida.

* Energia Refletida ${E_r}$: A energia refletida pode ser dividida em três outros processos de interação:
  1. Refração, quando ocorre amudança na direção do feixe de radiação (desvio)
devido a mudança de densidade – alterando a velocidade de propagação.
  2. Espalhamento: redistribuição da energia no espaço 3D
  3. Reflexão: retorno da energia ao meio original no qual ela se propagava
* Energia Absorvida ${E_a}$: Energia eletromagnética que incide sobre um corpo é convertida em outra forma de energia (Absorção)
* Energia Transmitida ${E_t}$: Energia eletromagnética que se propaga através do corpo (Transmissão)

---

*Processos de interação da energia*
\begin{equation}
{E_{inc,λ}} = {E_{r,λ}} + {E_{a,λ}} + {E_{t,λ}}
\end{equation}

Onde:
* ${E_{inc,λ}}$ = Energia incidida
* ${E_{r,λ}}$ = Energia Refletida
* ${E_{a,λ}}$ = Energia absorvida
* ${E_{t,λ}}$ = Energia transmitida

* ${r_λ} + {a_λ} + {t_λ} = 1$



#### Absortância / Absortividade ($α$)

É a fração da energia incidida que foi absorvida.

---

*Absortância*
\begin{equation}
α = \frac{E_{a,λ}}{E_{inc,λ}}
\end{equation}

Onde:
* $α$ = Absortância
* ${E_{a,λ}}$ = Energia absorvida
* ${E_{inc,λ}}$ = Energia incidida


#### Refletância / refletividade (r)

É a fração da energia incidida que foi refletida.

---

*Refletância*
\begin{equation}
r = \frac{E_{r,λ}}{E_{inc,λ}}
\end{equation}

Onde:
* $r$ = Refletância
* ${E_{r,λ}}$ = Energia refletida
* ${E_{inc,λ}}$ = Energia incidida

#### Transmitância / transmissividade (t)

É a fração da energia incidida que foi transmitida.

---

*Transmitância*
\begin{equation}
t = \frac{E_{t,λ}}{E_{inc,λ}}
\end{equation}

Onde:
* $t$ = Transmitância
* ${E_{t,λ}}$ = Energia transmitida
* ${E_{inc,λ}}$ = Energia incidida

#### Atenuação/extinção da radiação


A radiação é absorvida ou refletida dependendo do comprimento de onda, da energia que está sendo transportada, da distância optica/geométrica e da natureza do material que intervém, ou seja, da composição da matéria (tipo) e da sua concentração.

Para descrever a interação com a matéria é proposto uma variável ${β_{e,λ}}$ que indica a propriedade de atenuação/extinção da radiação, sendo este a soma dos coeficientes lineares de absorção ${β_{a,λ}}$ e espalhamento ${β_{s,λ}}$.

---

*Extinção/atenuação da radiação*
\begin{equation}
{L_λ}({S_2})-{L_λ}({S_1}) = -d{L_λ} = {β_{e, λ}}{L_λ}({S_1})ds
\end{equation}

Onde:
* $d{L_λ}$ = Energia extinta/atenuada pelo material
* ${β_{e, λ}}$ = Coeficientre de extinção/atenuação = ${β_{a,λ}}$+${β_{s,λ}}$
* ${L_λ}({S_1})$ = Total de energia incidida no material (energia incidida)
* ${L_λ}({S_2})$ = Total de energia após interação com o material (energia transmitida)
* $ds$ = Distância óptica/geométrica percorrida pela radiação no material





#### Coeficiente de extinção (${β_{e, λ}}$)

Reescrevendo a função anterior isolando o coeficiente ${β_{e, λ}}$ obtém uma função que pode ser interpretada como a taxa de redução de radiância ($\frac{{dL_λ}}{L_λ({S_1})}$) por unidade de caminho óptico ($ds$)

---

*Coeficiente de extinção*
\begin{equation}
{β_{e, λ}} = -\frac{(\frac{d{L_λ}}{{L_λ({S_1})}})}{ds}
\end{equation}

Onde:
* ${β_{e, λ}}$ = Coeficientre de extinção/atenuação = ${β_{a,λ}}$+${β_{s,λ}}$ (m-1)
* $-d{L_λ}$ = Energia extinta/atenuada pelo material
* ${L_λ({S_1})}$ = Total de energia incidida no material
* $ds$ = Distância óptica/geométrica percorrida pela radiação no material

```python
def coef_extincao(energExt, energInc, dist):
  return (energExt/energInc)/dist

print(str(coef_extincao((100-70), 100, 10))+" m^-1")
print(str(coef_extincao((100-57.58), 100, 14.14))+" m^-1")
```

#### Radiação transmitida (${L_λ({S_2})}$)

É possível obter a quantidade de radiação trasmitida ${L_λ({S_2})}$ integrando a equação anterior em $[{S_1}, {S_2}]$

---

\begin{equation}
{β_{e, λ}} = -\frac{(\frac{d{L_λ}}{{L_λ({S_1})}})}{ds}
\end{equation}

---
\begin{equation}
\frac{d{L_λ}}{{L_λ({S_1})}} = -{β_{e, λ}}{ds}
\end{equation}

---
\begin{equation}
\int_{S_1}^{S_2}\frac{d{L_λ}}{{L_λ({S_1})}} = -\int_{S_1}^{S_2}{β_{e, λ}}{ds}
\end{equation}

---
\begin{equation}
\ln(\frac{{L_λ({S_2})}}{{L_λ({S_1})}}) = -\int_{S_1}^{S_2}{β_{e, λ}}{ds}
\end{equation}

---
\begin{equation}
\exp[\ln(\frac{{L_λ({S_2})}}{{L_λ({S_1})}})] = \exp[-\int_{S_1}^{S_2}{β_{e, λ}}{ds}]
\end{equation}

---
\begin{equation}
\frac{{L_λ({S_2})}}{{L_λ({S_1})}} = \exp[-\int_{S_1}^{S_2}{β_{e, λ}}{ds}]
\end{equation}

---
\begin{equation}
{L_λ({S_2})} = {L_λ({S_1})}\exp[-\int_{S_1}^{S_2}{β_{e, λ}}{ds}]
\end{equation}

Onde:
* ${β_{e, λ}}$ = Coeficientre de extinção/atenuação = ${β_{a,λ}}$+${β_{s,λ}}$ (m-1)
* $-d{L_λ}$ = Energia extinta/atenuada pelo material
* ${L_λ({S_1})}$ = Total de energia incidida no material
* ${L_λ}({S_2})$ = Total de energia após interação com o material (energia transmitida)
* $ds$ = Distância óptica/geométrica percorrida pela radiação no material


#### Albedo de espalhamento simples ${ω_λ}$

Para melhor caracterizar a importância relativa do espalhamento versus a absorção, especialistas introduziram uma quantidade denominada "single
scattering albedo" (albedo de espalhamento simples) que varia de 0 a 1.

---

*Albedo de espalhamento simples*
\begin{equation}
{ω_λ} = \frac{β_{s,λ}}{β_{e,λ}} = \frac{β_{s,λ}}{{β_{a,λ}} + {β_{s,λ}}}
\end{equation}

Onde:
* ${ω_λ}$ = Albedo de espalhamento simples
* ${β_{e,λ}}$ = Coeficientre de extinção/atenuação
* ${β_{a,λ}}$ = Coeficientre de absorção
* ${β_{s,λ}}$ = Coeficientre de espalhamento

Casos de $ω, βa, βs$:

Material que                    | $ω$     |$βa$     | $βs$    | $βe$
--------------------------------|---------|---------|---------|---------------
Não interage                    |0        |0        |0        |βa=βs=0
Sem absorção                    |1        |0        |>0       |βe=0
Sem espalhamento                |0        |>0       |0        |βe=βs
1/2 absorvida e 1/2 espalhada   |1/2      |>0       |>0       |βa=βs; βe=βa+βs


#### É possível estimar a extinção do meio utilizando outro coeficiente ponderado pela quantidade do material?

#### Seção de choque(${σ_{e,λ}}$ )

Além do coeficiente de extinção é possível descrever a extinção no meio utilizando outro coeficiente ponderado pela quantidade do material (número de partículas por cm-3)

Exemplo: Aerossóis

Unidade de medida: m² ou cm²

#### Coeficiente	mássico	de	extinção (${\kappa_{e,λ}}$ )

Também é possível descrever a extinção no meio utilizando outro coeficiente ponderado pela quantidade do material (massa do material por m-3, densidade)

Exemplo : Gases distribuidos verticalmente como vapor d'água e Ozônio

Unidade de medida: m²/kg ou cm²/g

#### Mistura de componentes

\begin{equation}
β_{e, λ} = \sum_{i}^{n}{β_{e, λ, i}} = \sum_{i}^{n}{\rho_i\kappa_{e, λ, i}} = \sum_{i}^{n}{N_i\sigma_{e, λ, i}}
\end{equation}

##Espalhamento por moléculas e partículas atmosféricas#

Em algumas regiões do espectro solar o espalhamento reduz significativamente a radiação solar incidente à superfície.
Dentre as Constituintes espalhadores de radiação:
* moléculas gasosas,
* poeira,
* aerossóis,
* nuvens, etc.

Alguns fenômenos ópticos são causados pelo espalhamento edependem basicamente do tamanho da partícula espalhadora, dentre eles:
* arco íris,
* corona (anéis coloridos ao redor da lua),
* glory (sombra de objetos com anéis coloridos ao redor)
* miragem (refração da luz com meios diferentes indices de refração)

O espalhamento segue dois caminhos geometricos: Ao incidir em uma partícula a onda pode propagar no sentido contrário a incidencia ou retro-espalhar ou propagar-se no mesmo sentido da incidencia, espalhamento para frente

Pode tambem ser de 2 tipos, espalhamento simples, no qual os fótos são espalhados uma única vez, geralmente ocorre em meios oticamente dinos, ou espalhamento múltiplo que ocorrem em meios opticamente espessos, no qual os fótons podem se espalhar milhares de vezes antes de emergir do meio.

O espalhamento é determinado por 3 parâmetros:

* o comprimento de onda da radiação incidente

* o indíce de refração: propriedades ópticas da partícula relativas ao meio circunvizinho

* o tamanho da particula espalhadora, usualmente expressa no parâmetro adimensional, ${x}$:

\begin{equation}
x = \frac{2\pi{r}}{\lambda}
\end{equation}
onde:

${r}$: raio da partícula

$\lambda$: comprimento de onda



#### Regimes de espalhamento

Espalhamento Rayleigh: Partículas muito menores que os
λ da luz incidente.
O espalhamento é praticamente igual em todos as direções (Simétrica)

\begin{equation}
  \begin{cases}
    x << 1\\
    r << λ &
  \end{cases}
\end{equation}

Espalhamento Mie: Espalhamento de partículas que apresentam tamanho do comprimento de onda da radiação incidente. Função de fase – predomínio do espalhamento para frente

\begin{equation}
  \begin{cases}
    x \tilde= 1\\
    r \tilde= λ &
  \end{cases}
\end{equation}

Espalhamento óptica geométrica: Partículas maiores do que o comprimento de onda da radiação incidente. Depende do meio e da rugosidade do meio

\begin{equation}
  \begin{cases}
    x >> 1\\
    r >> λ &
  \end{cases}
\end{equation}


### Absorção e Emissão Seletiva

A atmosfera não é um corpo negro perfeito.

A atmosfera absorve radiação em alguns comprimentos de ondas, e pode ser transparente a outros comprimentos (radiação solar).

A atmosfera absorve pouca (5%) radiação nos comprimentos de ondas do visível.

O ozônio estratosférico absorve quase toda a radiação UV.

Vapor de água e dióxido de carbono são fortes absorvedores de radiação infravermelha e pobres absorvedor de radiação solar visível.

### Efeito Estufa

Os gases de efeito estufa já presentes na atmosfera muito antes das atividades humanas começarem a exercer influências na composição química da atmosfera.

• Exemplo: H2O, metano (CH4), óxido nitroso (N2O) e ozônio (O3).

O principal mecânismo pelo qual estes gases absorvem a radiação infravermelha é
através das rotação / vibrações de suas moléculas.

Se a atmosfera não tivesse os gases de efeito estufa, a temperatura média global da superfície da Terra seria aproximadamente –18° Celsius (255 K)

Gases de efeito estufa absorvem e emitem radiação no infravermelho (IR, também chamada de onda longa).

Parte desta radiação é emitida no sentido para cima e para baixo, aquecendo a superfície.

### Transições radiativas

* Rotacional: As transições radiativas puramente rotacional requerem que a molécula tenha um momento de dipolo permanente. Ex. (CO, N2O, H2O e O3).

* Vibracional-rotacional: Algumas moléculas pode não apresentam momento de dipolo permanente mas podem adquirir momento de dipolo induzido pelo modo vibracional. Ex. (CO2 e CH4)

* Vibracional: Se as cargas são distribuídas simetricamente, não há dipolo permanente e não há transição rotacional. Ex. (N2 e O2)

### Nuvens e Radiação

Nuvem = reflete a luz visível, absorve a radiação IR

* Não ocorre completo cancelamento do efeito radiativo na VIS e IR, depende do tipo de nuvens e estação do ano.

* Mudanças nas propriedades das nuvens, tipo e cobertura têm o potencial de afetar o clima.

* Nuvens altas geralmente são menos densas e permitem que mais radiação solar atravesse causando o aquecimento da nuvem, já as nuvens baixas ou espessas são mais densas e refletem mais a radiação causando um resfriamento da nuvem, por isso também tendem a aquecer a superficie a noite devido a reflexão da radiação de onda longa

* Os aerossois presentes na nuvem também afetam sua temperatura uma vez que esses podem apresentar um comportamento espalhador como Nitrato, sulfato e orgânicos, ou comportamento absorvedor como aerosol de poeira, black carbon, e outros.

### Forçante radiativa

Forçamento radiativa é geralmente definida como a alteração no balanço de radiação em um nível especifico da atmosfera devido ao efeito de uma forçante climática.

*  Esta quantidade é geralmente expressa em W/m²

* Forçante positiva => aquecimento;

* Forçante negativa => resfriamento.

### Integração espectral da TR no IR

#### Modelos Linha-a-linha

A integração linha-a-linha considera todas as linhas de absorção/emissão (continnum) conhecidas dentro do intervalo de integração considerado.

#### Integração espectral (modelos de banda)

Na prática, há interesse no fluxo calculado ou medido para todo o espectro.

* Então, é necessário obter valores de fluxo em banda larga com precisão razoável, mas com custo computacional bastante reduzido.

Alguns métodos possíveis para resolver integração espectral são os modelos de banda:

* Banda estreita (transmissão) – intervalos espectrais $Δv < 100 cm-1$ (menos eficiente computacionalmente mas é mais preciso)

* Banda larga (emissão) - intervalos espectrais $Δv > 100 cm-1$ (simplificação mais radical – rápida integração)

Os modelos mais comuns são:
* δ-distribution
* Expnencial
* Godson
* Malkmus


#### Método k-distribution

Os métodos k-distribution e correlated-k exploram o fato de que o mesmo valor de kv ocorre várias vezes em um intervalo espectral Δν.
* Essa redundância permite diminuir a quantidade de
cálculo da transmitância, realizando-o uma única vez!
* Os erros provenientes do uso do método da distribuição k
correlacionada nos valores de fluxos e de taxas de
aquecimento ficam em torno de 1%.
